```python
# Import necessary libraries

import pandas as pd
```

```python
# Set the Pandas option to display all columns

pd.set_option("display.max_columns", None)
```

```python
# Load the Airbnb dataset into a DataFrame

listing = pd.read_csv('listings.csv.gz', compression='gzip', header=0, sep=',', quotechar='"')
```

```python
# Display the first few rows of the dataset to understand its structure

listing.head()
```

## Get information from data

```python
# Display the column names of the DataFrame 'listing'

listing.columns
```

```python
# Define a list of columns to keep in the DataFrame

col = ["id", "neighbourhood_group_cleansed", "room_type", "accommodates", "bedrooms", "bathrooms_text", "minimum_nights", "maximum_nights", "price"]

# Create a new DataFrame 'data' that includes only the selected columns

data = listing[col]
```

```python
# Display the data types of each column

data.info()
```

```python
# Statistical summary of the dataset

data.describe()
```

```python
# Check for missing values in the dataset

data.isnull().sum() 
```

```python
# Data cleaning: Drop rows or columns with missing values if necessary

data = data.dropna()
data = data.reset_index(drop=True)
```

```python
#Clean price column, remove $ and convert it to float type

data["price"] = (data["price"].str.strip("$").str.replace(",", "")).astype(float)
```

```python
#Convert bath_text column to two columns

data[["bath", "bath_type"]] = data["bathrooms_text"].str.strip(" bath").str.strip(" baths").str.split(" ", expand = True)
```

```python
data[data.bath.apply(len) > 4]
```

```python
# Drop rows in the 'bath' column where the length is more than 4, as these are not numbers.

data = data[data.bath.apply(len) < 5]
data = data.reset_index(drop=True)
```

```python
# Convert the datatype of the 'bath' column to float

data["bath"] = data.bath.astype("float")
```

```python
# Create a temporary DataFrame 'tmp' that contains only the rows where the 'bath_type' column has missing values (NaN)

tmp = data[data.bath_type.isnull()].copy()

# Create a new DataFrame 'tmp2' by dropping all rows with any missing values (NaN) from the 'data' DataFrame

tmp2 = data.dropna()
```

```python
def room_bath(X):
    """
    Categorize room types into 'Shared' or 'Private'.

    Parameters:
    X (str): The room type to categorize. Expected values are typically 'Private room' or other types.

    Returns:
    str: 'Shared' if the room type is 'Private room'; otherwise, 'Private'.
    """
    if X == "Private room":
        return "Shared"
    
    return "Private"
```

```python
# Apply the 'room_bath' function to the 'room_type' column of the 'tmp' DataFrame

tmp.loc[:, "bath_type"] = tmp["room_type"].apply(room_bath)
```

```python
# Combine two DataFrames, 'tmp' and 'tmp2', into a single DataFrame

dfs = [tmp, tmp2]
data = pd.concat(dfs).reset_index(drop=True)
data = data.drop(columns=["bathrooms_text"])
data.head()
```

```python
# Check for missing values in the dataset

data.isnull().sum()
```

```python
# Display the data types of each column

data.info()
```

```python
# Statistical summary of the dataset

data.describe()
```

```python
# Save the DataFrame

data.to_csv("cleaned.csv", index=False)
```
